```python
# Read and print the CSV file in tabular form using loops
csv_filename = 'IRIS.csv'
lines=[]

with open(csv_filename, 'r') as csvfile:
    lines = csvfile.readlines()

#prints the header
row=lines[0].split(',')
print("{:<20} {:<20} {:<20} {:<20} {:<20}".format(row[0],row[1],row[2],row[3],row[4]))
print('-' * 100)

#prints rest of the file
for i in range(1,len(lines)):
  row = lines[i].split(',')
  print("{:<20} {:<20} {:<20} {:<20} {:<20}".format(row[0],row[1],row[2],row[3],row[4]))

```
